```python
import cv2
import matplotlib.pyplot as plt

image=cv2.imread('/content/drive/MyDrive/pavel-moiseev-8w3hXgLiL8w-unsplash.jpg')
rgb_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

```python
resized_image=cv2.resize(image,(300,300))
plt.imshow(cv2.cvtColor(resized_image,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.show()
```

```python
edges=cv2.Canny(gray_image,100,200)
plt.imshow(edges,cmap="gray")
plt.axis("off")
plt.show()
```

```python
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
faces=face_cascade.detectMultiScale(gray_image,scaleFactor=1.1,minNeighbors=5)
for(x,y,w,h) in faces:
  cv2.rectangle(rgb_image,(x,y),(x+w,y+h),(255,0,0),2)

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show
```

```python
contour,_=cv2.findContours(edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(rgb_image,contour,-1,(0,255,0),2)
plt.imshow(rgb_image)
plt.axis("off")
plt.show
```
